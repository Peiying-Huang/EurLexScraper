The DocumentSumScraper extract the following structure and raises an error when the website doesn't exist.\
The data structure:
1. 'Title and reference': string
2. 'Dates':[dict1, dict2]
3. 'Classifications':[dict1, dict2,...]
4. 'Summarised and linked documents':[dict1, dict2,...] (if the text is a hyperlink, the text and the link will be stored as a dictnary as the value of dictN)
5. 'Miscellaneous information':[dict1, dict2,...] (if the text is a hyperlink, the text and the link will be stored as a dictnary as the value of dictN)
6. 'Summary document url': a html link of English text  (most of the document summary websites contains a html form of the summarized text)
